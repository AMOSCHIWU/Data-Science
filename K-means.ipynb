```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import operator
import math as m
stress_data = pd.read_csv('work1.csv')

stress_data_w = stress_data.iloc[:,0:1].values
stress_data_h = stress_data.iloc[:,1:2].values
stress_data_y = stress_data.iloc[:,2:3].values
#n = stress_data_x.shape[0]
#c = stress_data_x.shape[1]
u1_w= list(stress_data_w[55])
u1_h= list(stress_data_h[55])
u2_w= list(stress_data_w[30])
u2_h= list(stress_data_h[30])
u3_w= list(stress_data_w[29])
u3_h= list(stress_data_h[29])
#print(stress_data_w[1])
#print(u1_w)
cluster = []
u1_w_update = 0
u1_h_update = 0
u2_w_update = 0
u2_h_update = 0
u3_w_update = 0
u3_h_update = 0
count_u1 = 0
count_u2 = 0
count_u3 = 0

cluster2 = []
time = 0
while True:
    plt_u1_w = []
    plt_u1_h = []
    plt_u2_w = []
    plt_u2_h = []
    plt_u3_w = []
    plt_u3_h = []
    for i in range(399):
        d1 = m.sqrt(np.square(stress_data_w[i] - u1_w ) + np.square(stress_data_h[i] - u1_h ))
        d2 = m.sqrt(np.square(stress_data_w[i] - u2_w ) + np.square(stress_data_h[i] - u2_h ))
        d3 = m.sqrt(np.square(stress_data_w[i] - u3_w ) + np.square(stress_data_h[i] - u3_h ))
        cluster.append([d1,d2,d3].index(min([d1,d2,d3])))  #取出計算後屬於哪一群
    
        if cluster[i] == 0:
            count_u1 += 1
            u1_w_update += stress_data_w[i]
            u1_h_update += stress_data_h[i]
            plt_u1_w.append(stress_data_w[i])
            plt_u1_h.append(stress_data_h[i])
        elif cluster[i] == 1:
            count_u2 += 1
            u2_w_update += stress_data_w[i]
            u2_h_update += stress_data_h[i]
            plt_u2_w.append(stress_data_w[i])
            plt_u2_h.append(stress_data_h[i])
        elif cluster[i] == 2:
            count_u3 += 1
            u3_w_update += stress_data_w[i]
            u3_h_update += stress_data_h[i]
            plt_u3_w.append(stress_data_w[i])
            plt_u3_h.append(stress_data_h[i])
    u1_w = u1_w_update / count_u1
    u1_h = u1_h_update / count_u1
    u2_w = u2_w_update / count_u2
    u2_h = u2_h_update / count_u2
    u3_w = u3_w_update / count_u3
    u3_h = u3_h_update / count_u3
    time = time + 1
    if operator.eq(cluster,cluster2):
        plt.scatter(plt_u1_w,plt_u1_h,c = 'yellow',label='class_1')
        plt.scatter(plt_u2_w,plt_u2_h,c = 'lightgray',label='class_2')
        plt.scatter(plt_u3_w,plt_u3_h,c = 'pink',label='class_3')
        plt.scatter(u1_w,u1_h,c = 'red',label='class_1_mean')
        plt.scatter(u2_w,u2_h,c = 'blue',label='class_2_mean')
        plt.scatter(u3_w,u3_h,c = 'orange',label='class_3_mean')
        plt.title('K means')
        plt.legend(loc = 0)
        plt.show()
        #print(u1_w,u1_h)
        break
    else:
        #print(u2_w,u2_h)
        u1_w_update = 0
        u1_h_update = 0
        u2_w_update = 0
        u2_h_update = 0
        u3_w_update = 0
        u3_h_update = 0
        count_u1 = 0
        count_u2 = 0
        count_u3 = 0
        cluster2 = cluster
        #print(cluster2)
        
        #print(cluster)
        cluster = []
        #print(cluster2)
        #print(cluster)
```
